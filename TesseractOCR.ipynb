```python
from PIL import Image
import pytesseract
from google.colab import files
import os

# Install Tesseract OCR and the Bengali language package
!sudo apt-get update
!sudo apt-get install tesseract-ocr
!sudo apt-get install tesseract-ocr-ben  # Install Bengali language pack

# Verify the installation of the Bengali language data
!ls /usr/share/tesseract-ocr/4.00/tessdata/

# Set the correct path for Tesseract in Colab
pytesseract.pytesseract.tesseract_cmd = '/usr/bin/tesseract'

# Language configuration for Bangla
lang = 'ben'  # 'ben' stands for Bengali

# Function to extract Bangla text from an uploaded image
def extract_bangla_text():
    # Upload the image using Colab's file upload
    uploaded = files.upload()

    # Get the file name from the uploaded files
    for filename in uploaded.keys():
        print(f"Processing file: {filename}")

        # Open the image using PIL
        image = Image.open(filename)

        # Use pytesseract to extract text in Bangla
        text = pytesseract.image_to_string(image, lang=lang)

        # Print the extracted text
        print("Extracted Bangla Text:")
        print(text)

# Call the function to upload and extract text
extract_bangla_text()
```

```python
# Import necessary libraries
import pytesseract
from PIL import Image
import re
from google.colab import files
import nltk
nltk.download('stopwords')

# Define Bengali stopwords (You can add more words to this list)
bengali_stopwords = set([
    "এটি", "এবং", "আপনি", "আমি", "এই", "সে", "তাদের", "এরা", "একটি", "তবে", "কিছু", "করুন",
    "থেকে", "যত", "কেন", "কেননা", "হবে", "ছিল", "হয়", "এমন", "যারা", "সেই", "তিন"
    # Add more Bengali stopwords here as necessary
])

# Language configuration for Bangla
lang = 'ben'  # 'ben' stands for Bengali

# Function to clean extracted Bangla text
def clean_bengali_text(text):
    # Keep only Bengali characters, remove non-Bengali characters (including emojis, numbers, etc.)
    text = re.sub(r'[^\u0980-\u09FF\s]', '', text)  # Retain only Bengali characters (Unicode range: U+0980 to U+09FF)

    # Remove Bengali stopwords
    text = ' '.join([word for word in text.split() if word not in bengali_stopwords])

    # Remove extra spaces and newlines
    text = re.sub(r'\s+', ' ', text).strip()  # Replace multiple spaces with single space and remove leading/trailing spaces

    # Convert text to lowercase
    text = text.lower()

    return text

# Function to upload the image, extract text, and clean it
def extract_and_clean_bangla_text():
    # Upload the image using Colab's file upload
    uploaded = files.upload()

    # Get the file name from the uploaded files
    for filename in uploaded.keys():
        print(f"Processing file: {filename}")

        # Open the image using PIL
        image = Image.open(filename)

        # Use pytesseract to extract text in Bangla
        text = pytesseract.image_to_string(image, lang=lang)

        # Print the extracted text
        print("Extracted Bangla Text:")
        print(text)

        # Clean the extracted text
        cleaned_text = clean_bengali_text(text)

        # Print the cleaned text
        print("Cleaned Bangla Text:")
        print(cleaned_text)

# Call the function to upload, extract, and clean text
extract_and_clean_bangla_text()
```
